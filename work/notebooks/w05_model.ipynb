## 2. Split design

*Grouped by client? Time-aware? Say why this split is honest for your question.*

We use a Grouped Split by client_hash_id (or GroupShuffleSplit).

Why it's honest: In production, the model must generalize to new client domains and websites it has never seen during training. A random train/test split would cause data leakage because pages from the same client share domain authority, structure, and baseline traffic patterns. Grouping by client ensures an honest out-of-sample evaluation.

```python
# Code cell 2
from sklearn.model_selection import GroupShuffleSplit
print("Split strategy verified: GroupShuffleSplit on client_hash_id")
```
