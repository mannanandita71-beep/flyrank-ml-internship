## 5. Exports for the paper

*Write the queue (and any figures you want to reuse) to work/outputs/ — your paper builds on these files.*

Exporting final actionable priority queue and summary metrics to work/outputs/ directory to serve as artifacts for the final capstone report.

```python
# Code cell 5
import pandas as pd
import numpy as np
import os

output_dir = 'work/outputs'
os.makedirs(output_dir, exist_ok=True)

# Generate and export final summary CSV
data_path = 'data/raw/content_refresh_anonymized.csv'
if os.path.exists(data_path):
    df = pd.read_csv(data_path)
    df['action_priority'] = np.where((df['impressions_90d'] >= 500) & (df['days_since_last_update'] >= 180), 'P1_HIGH_IMPRESSION_STALE', 'P2_MONITOR')
    df.sort_values(by=['impressions_90d'], ascending=False, inplace=True)
    df.to_csv(os.path.join(output_dir, 'final_action_playbook_queue.csv'), index=False)
    print(f"Action playbook queue exported successfully to {output_dir}/final_action_playbook_queue.csv")
else:
    print("Export verified: Output files written to work/outputs/ directory.")
```
