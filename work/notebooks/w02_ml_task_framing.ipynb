## 3. Success metric

*One metric you can defend. What number means 'good'?*

Metric: Precision@50

What means 'good': Precision@50 >= 0.70 (70% or higher).

Defense: Content strategists usually review a limited batch (e.g., top 50 pages) per operational cycle. Precision@50 directly reflects how many of those top 50 recommended pages were truly declining and worth editing.

```python
# Code cell 3
target_precision = 0.70
print(f"Primary evaluation metric: Precision@50 (Target: >= {target_precision:.0%})")
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A fixed if-statement rule (e.g., impressions > 500 AND days > 180) is too rigid and misses non-linear interactions between multiple features like position fluctuations, click-through rate variations, and content age. Machine learning models adaptively weight these complex interactions simultaneously without hardcoding arbitrary thresholds.
